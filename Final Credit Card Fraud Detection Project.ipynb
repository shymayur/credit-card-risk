### 1. Importing dependencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

### 2. Importing dataset, checking information, shape & Null-values

```python
credit_card_data=pd.read_csv('creditcard.csv')
credit_card_data.head()
```

```python
credit_card_data.info()
```

```python
credit_card_data.shape
```

```python
credit_card_data.isnull().sum()
```

### 3. Distribution of legit & fraud transactions

```python
credit_card_data['Class'].value_counts()
```

#### 0-----> Legit
#### 1-----> Fraud

### 4. Balanced or Unbalanced check

```python
credit_card_data.groupby('Class').mean()
```

```python
legit = credit_card_data[credit_card_data['Class']==0]
fraud = credit_card_data[credit_card_data['Class']==1]

print(legit.shape)
print(fraud.shape)
```

#### The distribution shows that this dataset is highly un-balanced

```python
credit_card_data.describe()
```

```python
legit.describe()
```

```python
fraud.describe()
```

### 5. Creating balanced dataset
#### Build a sample dataset containing similar distribution of Legit transactions and Fraud transactions (both equal to 492)

```python
legit_sample = legit.sample(n=492)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

#### Now this "new_dataset" has balanced data.

```python
new_dataset.groupby('Class').mean()
```

### 6. Splitting the dataset into Features (X) & Targets (Y)

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
X
```

```python
Y
```

### 7. Split the data into Training data & Testing Data

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, x_train.shape, x_test.shape)
```

```python
print(Y.shape, y_train.shape, y_test.shape)
```

### 8. Model training : Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

### 9. Model evaluation on training data

```python
x_train_predicted = model.predict(x_train)
train_accuracy = accuracy_score(x_train_predicted, y_train)
train_accuracy
```

### 10. Model evaluation on test data

```python
x_test_predicted = model.predict(x_test)
test_accuracy = accuracy_score(x_test_predicted, y_test)
test_accuracy
```

### 11. Precision, Recall and F1 score

```python
report = classification_report(x_test_predicted, y_test)
print(report)
```
